# Problem 11

```python
def sum(arr, n): 
  
    i = n 
    if(arr[i] < 9): 
      
        arr[i] = arr[i] + 1
        return
    arr[i] = 0
    i -= 1
    sum(arr, i) 
      
n = 4
arr = [0, 1, 9, 3, 9] 
  
sum(arr, n) 
if(arr[0] > 0): 
    print(arr[0], ", ", end="") 
      
for i in range(1, n+1): 
      
    print(arr[i], end="") 
    if(i < n): 
        print(", ", end="") 
```

# Problem 12

```python
a = np.arange(15).reshape(5, 3)
b = np.arange(9).reshape(3, 3)
print(np.diag(np.dot(np.dot(a, b), a.T)))
print(np.einsum('ij,ji->i', np.dot(a, b), a.T))
print(np.einsum('ij,ij->i', np.dot(a, b), a))
```
